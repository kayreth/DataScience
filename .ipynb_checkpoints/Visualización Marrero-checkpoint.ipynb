```python
#Importar Librerias
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Subir archivo
Activaciones_df = pd.read_csv('Reporte Activaciones.csv', delimiter=';', encoding='latin1')
```

```python
Activaciones_df
```

```python
#Nombres de las columnas
print(Activaciones_df.columns)
```

```python
#Contar la cantidad de veces que se repite cada valor en la columna 'TaxID Proveedor'
proveedor_counts = Activaciones_df['TaxID Proveedor'].value_counts()
```

```python
# Seleccionar los 15 primeros TaxID de proveedor más comunes
top_proveedores = proveedor_counts.head(15)
```

```python
# Obtener los nombres de los proveedores correspondientes a los TaxID
nombres_proveedores = Activaciones_df.groupby('TaxID Proveedor')['Nombre Proveedor'].first()
nombres_proveedores_top = nombres_proveedores.loc[top_proveedores.index]
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
top_proveedores.plot(kind='bar', color='skyblue')
plt.title('Top 15 de TaxID de Proveedor más comunes con nombres de proveedor')
plt.xlabel('TaxID de Proveedor')
plt.ylabel('Cantidad de Repeticiones')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor visualización
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Agregar una cuadrícula horizontal
```

```python
# Crear el gráfico de barras con un tamaño más grande
plt.figure(figsize=(15, 8))
top_proveedores.plot(kind='bar', color='skyblue')
plt.title('Top 15 de TaxID de Proveedor más comunes con nombres de proveedor')
plt.xlabel('TaxID de Proveedor')
plt.ylabel('Cantidad de Repeticiones')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor visualización
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Agregar una cuadrícula horizontal

# Agregar nombres de proveedores como anotaciones en las barras
for i, proveedor_count in enumerate(top_proveedores):
    plt.text(i, proveedor_count + 0.2, f'{nombres_proveedores_top.iloc[i]}', ha='center')

# Ajustar manualmente los márgenes
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.3)

plt.show()
```

```python
# Crear etiquetas compuestas para el eje x con TaxID y nombre de proveedor
etiquetas_x = [f"{tax_id}\n{nombre}" for tax_id, nombre in zip(top_proveedores.index, nombres_proveedores_top)]

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
top_proveedores.plot(kind='bar', color='skyblue')
plt.title('Top 15 de TaxID de Proveedor más comunes con nombres de proveedor')
plt.ylabel('Cantidad de Repeticiones')
plt.xticks(range(len(etiquetas_x)), etiquetas_x, rotation=45, ha='right')  # Usar etiquetas compuestas en el eje x
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Agregar una cuadrícula horizontal

# Mostrar la cantidad de repeticiones en la parte superior de cada barra
for i, proveedor_count in enumerate(top_proveedores):
    plt.text(i, proveedor_count + 0.2, str(proveedor_count), ha='center')

plt.tight_layout()  # Ajustar el diseño del gráfico
plt.show()
```

```python
#Contar la cantidad de veces que se repite cada valor en la columna 'TaxID Comprador'
comprador_counts = Activaciones_df['TaxID Comprador'].value_counts()
```

```python
#Mostrar resultados
comprador_counts
```

```python
# Obtener los nombres de los compradores correspondientes a los TaxID
nombres_compradores = Activaciones_df.groupby('TaxID Comprador')['Nombre Comprador'].first()
```

```python
#Mostrar resultados
nombres_compradores
```

```python
# Crear una nueva tabla con TaxID de comprador, nombre de comprador y contador de números de cotización
nueva_tabla = pd.DataFrame({
    'TaxID Comprador': comprador_counts.index,
    'Nombre Comprador': nombres_compradores.loc[comprador_counts.index].values,
    'Contador N° Cotizacion': comprador_counts.values
})
nueva_tabla
```

```python
# Calcular el total de cotizaciones por cada comprador
total_por_comprador = nueva_tabla.groupby(['TaxID Comprador', 'Nombre Comprador'])['Contador N° Cotizacion'].sum()
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
ax = total_por_comprador.plot(kind='bar', color='skyblue')
plt.title('Cantidad de N° Cotizaciones por Comprador')
plt.xlabel('Comprador (TaxID - Nombre)')
plt.ylabel('Total de N° Cotizaciones')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor visualización
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Agregar una cuadrícula horizontal

# Agregar el total en cada barra
for i, valor in enumerate(total_por_comprador):
    ax.text(i, valor + 0.2, str(valor), ha='center')

plt.tight_layout()  # Ajustar el diseño del gráfico
plt.show()
```

```python
# Calcular el total de cotizaciones por cada comprador y ordenar de mayor a menor
total_por_comprador_sorted = total_por_comprador.sort_values(ascending=False)
```

```python
#ordenar de mayor a menor y eliminar los valores del eje y?
# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
ax = total_por_comprador_sorted.plot(kind='bar', color='skyblue')
plt.title('Cantidad de N° Cotizaciones por Comprador (ordenado)')
plt.xlabel('Comprador (TaxID - Nombre)')
plt.ylabel('Total de N° Cotizaciones')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor visualización
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Agregar una cuadrícula horizontal

# Ocultar los valores del eje y
plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)

# Agregar el total en cada barra
for i, valor in enumerate(total_por_comprador_sorted):
    ax.text(i, valor + 0.2, str(valor), ha='center')

plt.tight_layout()  # Ajustar el diseño del gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filtrar el DataFrame para el comprador con mayor número de cotizaciones
comprador_max_cotizaciones_data = Activaciones_df[Activaciones_df['TaxID Comprador'] == '6-00040629']

# Crear una figura y ejes para los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Contar la cantidad de cotizaciones en cada estado
conteo_visualizadas = comprador_max_cotizaciones_data['RFQ Visualizada'].value_counts()
conteo_aceptadas = comprador_max_cotizaciones_data['RFQ Aceptada'].value_counts()
conteo_rechazadas = comprador_max_cotizaciones_data['RFQ Rechazada'].value_counts()
conteo_ofertadas = comprador_max_cotizaciones_data['RFQ Ofertada'].value_counts()

# Colores para las barras
colores = ['lightblue', 'skyblue', 'dodgerblue', 'darkblue']

# Etiquetas para los estados de cotización
estados = ['Visualizadas', 'Aceptadas', 'Rechazadas', 'Ofertadas']

# Iterar sobre los subgráficos y los datos para graficar
for ax, conteo, color, estado in zip(axs.flatten(), [conteo_visualizadas, conteo_aceptadas, conteo_rechazadas, conteo_ofertadas], colores, estados):
    ax.bar(conteo.index, conteo.values, color=color)
    ax.set_title(f'Cantidad de N° Cotizaciones {estado}')
    ax.set_xlabel('Estado de Cotización')
    ax.set_ylabel('Cantidad de N° Cotizaciones')
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Agregar los totales en cada gráfico
    for i, valor in enumerate(conteo.values):
        ax.text(i, valor, str(valor), ha='center', va='bottom')

# Ajustar el diseño de los subgráficos
plt.tight_layout()
plt.show()
```

```python
#Otra forma
# Configuración de estilo de Seaborn
sns.set(style="whitegrid")

# Crear el gráfico de barras con Seaborn
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=Activaciones_df, x='RFQ Visualizada', palette='Blues', alpha=0.7, label='Visualizadas')
sns.countplot(data=Activaciones_df, x='RFQ Aceptada', palette='Greens', alpha=0.7, label='Aceptadas')
sns.countplot(data=Activaciones_df, x='RFQ Rechazada', palette='Oranges', alpha=0.7, label='Rechazadas')
sns.countplot(data=Activaciones_df, x='RFQ Ofertada', palette='Reds', alpha=0.7, label='Ofertadas')
plt.title('Cantidad de veces que cada estado de cotización aparece')
plt.xlabel('Estado de Cotización')
plt.ylabel('Cantidad de Cotizaciones')
plt.legend()

# Agregar valores en las barras
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.show()
```

```python
# Obtener los conteos de cotizaciones por source
conteo_por_source = Activaciones_df['source'].value_counts()

# Configurar los colores para las secciones del gráfico
colores = ['skyblue', 'lightgreen']

# Configurar las etiquetas para las secciones del gráfico
etiquetas = conteo_por_source.index

# Crear el gráfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(conteo_por_source, labels=etiquetas, autopct='%1.1f%%', colors=colores, startangle=140)
plt.title('Distribución de Cotizaciones por Source')
plt.axis('equal')  # Hacer que el gráfico sea circular

# Añadir los valores en el centro del gráfico
plt.annotate(f'Total: {conteo_por_source.sum()}', xy=(0, 0), fontsize=12, ha="center", va="center")

plt.show()
```

```python
# Filtrar los datos donde % Ofertas en RFQ es 0
datos_con_ofertas_cero = Activaciones_df[Activaciones_df['% Ofertas en RFQ'] == 0]

# Agrupar los datos por cotización y contar la frecuencia de cada estado de cotización
cotizaciones_estado = datos_con_ofertas_cero.groupby('N° Cotizacion')['RFQ Visualizada'].value_counts().unstack(fill_value=0)

# Obtener las 20 cotizaciones más frecuentes
cotizaciones_top20 = cotizaciones_estado.sum(axis=1).nlargest(20)

# Seleccionar las cotizaciones más frecuentes y sus estados de cotización correspondientes
cotizaciones_top20_estado = cotizaciones_estado.loc[cotizaciones_top20.index]

# Crear el gráfico de barras con Seaborn
plt.figure(figsize=(12, 6))
sns.scatterplot(data=cotizaciones_top20_estado.T, palette='Set2')
plt.title('Top 20 de N° Cotizacion donde % Ofertas en RFQ fue 0')
plt.xlabel('Estado de Cotización')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.grid(True)
plt.legend(title='N° Cotizacion')
plt.show()
```
