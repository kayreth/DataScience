```python
#Importar Librerias
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Subir el archivo Activaciones
Activaciones_df = pd.read_csv('Reporte Activaciones.csv', delimiter=';', encoding='latin1')
```

```python
#Visualizar datos del DataFrame
Activaciones_df
```

```python
#Identificar los nombres de las columnas del DataFrame
print(Activaciones_df.columns)
```

```python
#Verificar el tipo de datos de las columnas en el DataFrame 
Tipo_Datos = Activaciones_df.dtypes
Tipo_Datos
```

```python
#Contar la cantidad de proveedores diferentes y mostrar resultados
Total_Proveedores = Activaciones_df['TaxID Proveedor'].nunique()
print("Cantidad de proveedores diferentes: ",Total_Proveedores)
```

```python
#Contar la cantidad de compradores diferentes y mostrar resultados
Total_Compradores = Activaciones_df['TaxID Comprador'].nunique()
print("Cantidad de Compradores diferentes: ",Total_Compradores)
```

```python
#Contar la cantidad de RFQs y mostrar resultados
Total_RFQs = Activaciones_df['N° Cotizacion'].nunique()
print("Hay:",Total_RFQs ,"RFQs")
```

```python
#Convertir las columnas Fecha de Publicación, Fecha de creacion y Fecha de Cierre a formato de fecha
Activaciones_df['Fecha de Publicación'] = pd.to_datetime(Activaciones_df['Fecha de Publicación'])
Activaciones_df['Fecha de creacion'] = pd.to_datetime(Activaciones_df['Fecha de creacion'])
Activaciones_df['Fecha de Cierre'] = pd.to_datetime(Activaciones_df['Fecha de Cierre'])
```

```python
#Gráfico 1: Visualizar los proveedores que han sido invitados a participar en mas procesos de cotización, mediante
#la relación entre nombre del proveedor y la fecha de publicación.

#Agrupacion de los datos por TaxID de Proveedor y contar la cantidad de veces que se repite cada TaxID
Top15_Proveedores = Activaciones_df['TaxID Proveedor'].value_counts().head(15)

#Filtrar los datos para mostrar solo los 15 más comunes
Datos_Top15 = Activaciones_df[Activaciones_df['TaxID Proveedor'].isin(Top15_Proveedores.index)].copy()

#Extraer el mes y el año de la Fecha de Publicación y convertir a cadena
Datos_Top15.loc[:, 'Mes-Año'] = Datos_Top15['Fecha de Publicación'].dt.to_period('M').astype(str)

#Crear el gráfico
plt.figure(figsize=(12, 6))
for proveedor, datos_proveedor in Datos_Top15.groupby('TaxID Proveedor'):
    plt.plot(datos_proveedor['Mes-Año'], datos_proveedor['Nombre Proveedor'], marker='o', linestyle='', label=proveedor)
plt.title('Top 15 de Proveedor invitados a un N° de Cotizacion por Mes y Año de Publicación')
plt.xlabel('Mes y Año de Publicación')
plt.ylabel('Nombre Proveedor')
plt.xticks(rotation=45)
plt.legend(title='TaxID Proveedor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()


#Cada TaxID de Proveedor estará representado por una línea, y cada punto en la línea indicará la fecha de publicación en la 
#que el proveedor fue invitado a un proceso de cotización. En la leyenda se muestran los TaxID de Proveedor correspondientes.
```

```python
#Gráfico 2: Calcular el total de cotizaciones para cada empresa compradora por mes y año de publicación.

#Extraer mes y año de la Fecha de Publicación y convertir a cadena
Activaciones_df['Mes-Año'] = Activaciones_df['Fecha de Publicación'].dt.to_period('M').astype(str)

#Agrupar los datos por TaxID y Nombre de la empresa compradora y Mes-Año, y frecuencia de cotizaciones
Cotizaciones_Comprador = Activaciones_df.groupby(['TaxID Comprador', 'Nombre Comprador', 'Mes-Año']).size().reset_index(name='Total Cotizaciones')

#Crear el gráfico
plt.figure(figsize=(16, 8))
ax = Cotizaciones_Comprador.plot(x='Mes-Año', y='Total Cotizaciones', kind='bar', color='skyblue')
plt.title('Total de N° Cotizacion por Empresa Compradora y Mes, Año de su Publicación')
plt.xlabel('Mes-Año de Publicación')
plt.ylabel('Total de N° Cotización')
plt.xticks(rotation=45)
plt.grid(axis='y')

#Agregar los totales a cada una de las barras
for index, value in enumerate(Cotizaciones_Comprador['Total Cotizaciones']):
    ax.text(index, value, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

#Cada barra en el gráfico representa el total de cotizaciones para un cada empresa compradora en un mes y año específicos.
```

```python
#Gráfico 3: Agrupación los datos por fuente RFQ y cuentar la frecuencia de cada aceptación o rechazo de RFQ.

#Filtro donde RFQ Aceptada o RFQ Rechazada es SI
Rfq_Aceptada = Activaciones_df[Activaciones_df['RFQ Aceptada'] == 'SI']
Rfq_Rechazada = Activaciones_df[Activaciones_df['RFQ Rechazada'] == 'SI']

#Contar la frecuencia de RFQ Aceptada y Rechazada por fuentes
Conteo_Aceptada = Rfq_Aceptada['source'].value_counts()
Conteo_Rechazada = Rfq_Rechazada['source'].value_counts()

#Combinar los conteos en un DataFrame
Cuenta_Total = pd.concat([Conteo_Aceptada, Conteo_Rechazada], axis=1, keys=['RFQ Aceptada', 'RFQ Rechazada']).fillna(0)

#Crear el gráfico
plt.figure(figsize=(10, 6))
Cuenta_Total.plot(kind='bar', stacked=True, colormap='Set2', ax=plt.gca())
plt.title('Total de veces que RFQ fueron Aceptadas y Rechazadas por fuente')
plt.xlabel('Fuente')
plt.ylabel('Totales')
plt.xticks(rotation=45)
plt.legend(title='Estado RFQ')
plt.grid(axis='y')

#Agregar los totales a cada una de las barras
for p in plt.gca().patches:
    plt.gca().annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()

#El gráfico, permite visualizar cómo se distribuyen las cotizaciones en dos estados de RFQ para cada fuente. 
#Cada barra representa el total de cotizaciones para un estado de RFQ específico, y las barras apiladas muestran la 
#contribución de cada estado de RFQ al total.
```

```python
#Gráfico 4: Combina las Fechas de creacion y Publicación en una sola serie de fechas, para extraer el día de cada fecha 
#y mostrar la cantidad de cotizaciones por día.

#Obtener las fechas relevantes, combinando fecha de creacion y fecha de publicación
Fechas = pd.concat([Activaciones_df['Fecha de creacion'], Activaciones_df['Fecha de Publicación']])

#Extraer el día de cada fecha
Fechas_Dia = Fechas.dt.day

#Crear el gráfico
plt.figure(figsize=(12, 8))
axi = sns.countplot(x=Fechas_Dia, palette='viridis')

#Agregar los totales a cada una de las barras
for p in axi.patches:
    axi.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 5), 
                textcoords = 'offset points')

plt.title('Cantidad de cotizaciones por día')
plt.xlabel('Día')
plt.ylabel('Cantidad de cotizaciones')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Este gráfico muestra los días del mes donde hay mayor cantidad de publicacion de RFQ por parte de las empresas compradoras.
```

```python
#Gráfico 5: Total de RFQs creadas por los compradores

#Crear el gráfico
plt.figure(figsize=(12, 8))
axi = sns.countplot(x='Usuario Comprador', hue='TaxID Comprador', data=Activaciones_df, palette='viridis')

#Agregar los totales a cada una de las barras excluyendo 0
for p in axi.patches:
    if p.get_height() > 0:
        axi.annotate(format(p.get_height(), '.0f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='center', 
                     xytext=(0, 5), 
                     textcoords='offset points')

plt.title('Cantidad de procesos de cada Usuario Comprador por N° Cotización y TaxID Comprador')
plt.xlabel('Usuario Comprador')
plt.ylabel('Cantidad de procesos')
plt.xticks(rotation=45)
plt.legend(title='TaxID Comprador', loc='upper right')
plt.tight_layout()
plt.show()

#El resultado es el total de procesos por cada empresa compradora, detallando todos los valores mayores que 0
```

```python
#Gráfico 6: Top 20 del total de cotizaciones por proveedor distintas de 0, donde la RFQ fue rechazada

#Filtrar solo las cotizaciones donde la RFQ Rechazada sea SI
Cotizaciones_Rechazada = Activaciones_df[Activaciones_df['RFQ Rechazada'] == 'SI']

#Contar la cantidad de cotizaciones por Proveedor
Cotizaciones_Proveedor = Cotizaciones_Rechazada['TaxID Proveedor'].value_counts().head(20)

#Obtener solo las cotizaciones de los proveedores en el top 20
Cotizaciones_TopProveedores = Cotizaciones_Rechazada[Cotizaciones_Rechazada['TaxID Proveedor'].isin(Cotizaciones_Proveedor.index)]

#Total de cotizaciones por proveedor
Total_Proveedor = Cotizaciones_TopProveedores.groupby('TaxID Proveedor').size()

#Filtrar los totales diferentes de 0
Total_Proveedor_Filtrado = Total_Proveedor[Total_Proveedor != 0]

#Ordenar los proveedores por el total de cotizaciones de mayor a menor
Total_Proveedor_Ordenado = Total_Proveedor_Filtrado.sort_values(ascending=False)

#Filtrar las cotizaciones por los proveedores ordenados
TopPO_Cotizaciones = Cotizaciones_TopProveedores[Cotizaciones_TopProveedores['TaxID Proveedor'].isin(Total_Proveedor_Ordenado.index)]

#Crear el gráfico
plt.figure(figsize=(16, 8))
ax = sns.countplot(x='TaxID Proveedor', hue='Nombre Proveedor', data=TopPO_Cotizaciones, order=Total_Proveedor_Ordenado.index, palette='viridis')
plt.title('Top 20 de Proveedores invitados a RFQ que han Rechazado el proceso')
plt.xlabel('TaxID Proveedor')
plt.ylabel('Cantidad de Cotizaciones')
plt.xticks(rotation=45)
plt.legend(title='Nombre Proveedor', loc='upper right')

# Agregar los totales como anotaciones en las barras
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(format(int(p.get_height()), '.0f'), 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 5), 
                    textcoords='offset points')

plt.tight_layout()
plt.show()

#El resultado es un total de veces que los proveedores han rechazado participar en diferentes procesos de cotización.
```
